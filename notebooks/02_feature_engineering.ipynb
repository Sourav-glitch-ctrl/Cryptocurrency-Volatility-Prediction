```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\soura\OneDrive\Desktop\Crypto Currency\data\processed\processed_data.csv")
data['date'] = pd.to_datetime(data['date'])
data = data.sort_values('date').reset_index(drop=True)
```

```python
data['target'] = (data['volatility'].shift(-1) > data['volatility']).astype(int)
```

```python
data['vol_lag1'] = data['volatility'].shift(1)
data['vol_lag2'] = data['volatility'].shift(2)
data['vol_lag3'] = data['volatility'].shift(3)
```

```python
data['ret_lag1'] = data['returns'].shift(1)
data['ret_lag2'] = data['returns'].shift(2)
data['ret_lag3'] = data['returns'].shift(3)
```

```python
data['vol_5']  = data['returns'].rolling(window=5).std()
data['vol_10'] = data['returns'].rolling(window=10).std()
data['vol_20'] = data['returns'].rolling(window=20).std()
```

```python
data['hl_diff'] = data['high'] - data['low']
data['co_diff'] = data['close'] - data['open']
```

```python
data['ma_5']  = data['close'].rolling(window=5).mean()
data['ma_10'] = data['close'].rolling(window=10).mean()
data['ma_20'] = data['close'].rolling(window=20).mean()
```

```python
data['ma_ratio'] = data['ma_5'] / data['ma_20']
```

```python
data['momentum_5'] = data['close'] - data['close'].shift(5)
```

```python
data['vol_of_vol'] = data['volatility'].rolling(window=5).std()
```

```python
# Trend strength
data['trend_10'] = data['close'] / data['ma_10']

# Volatility momentum
data['vol_change'] = data['volatility'] - data['volatility'].shift(1)

# Return volatility interaction
data['ret_vol_interaction'] = data['returns'] * data['volatility']

# Z-score normalization
data['zscore_vol'] = (data['volatility'] - data['vol_20']) / data['vol_20']

# Range expansion
data['range_pct'] = (data['high'] - data['low']) / data['close']
```

```python
# Momentum
data['momentum_5'] = data['close'].pct_change(5)

# Trend strength
data['trend_strength'] = data['ma_10'] - data['ma_20']

# Volatility regime
data['high_vol_regime'] = (data['volatility'] > data['vol_20']).astype(int)

# Return acceleration
data['return_acc'] = data['returns'] - data['returns'].shift(1)
```

```python
# Price momentum
data['momentum_3'] = data['close'].pct_change(3)

# Trend difference
data['ma_diff'] = data['ma_10'] - data['ma_20']

# Volatility breakout
data['vol_breakout'] = (data['volatility'] > data['vol_20']).astype(int)

# Return strength
data['return_strength'] = data['returns'].rolling(5).mean()
```

```python
# Target (regression)
data['target'] = data['volatility'].shift(-1)

# Rolling features (shifted)
data['vol_5'] = data['returns'].rolling(5).std().shift(1)
data['ma_5'] = data['close'].rolling(5).mean().shift(1)

# Momentum (already safe)
data['momentum_5'] = data['close'].shift(1) - data['close'].shift(6)

# Vol of vol
data['vol_of_vol'] = data['volatility'].rolling(5).std().shift(1)
```

```python
data = data.dropna().reset_index(drop=True)
```

```python
features = [col for col in data.columns if col not in ['target', 'date']]
```

```python
data = data.drop(columns=['Unnamed: 0'])
```
